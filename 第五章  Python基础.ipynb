```python
'''
类，这里写一个学生的类，了解类的写法
'''
class Student:
    
    student_count = 0
    
    def __init__(self, name, age):    
        self.name = name       
        self.age = age        
        Student.student_count += 1
    
    def dis_student(self):    
        print("Student name:", self.name, "Student age:", self.age)
    
student1 = Student("Tom", 10)
student2 = Student("Jarry", 10)
student1.dis_student()
student2.dis_student()
```

```python
'''
类的继承，当一个类被继承时，这个类的初试方法是不会被自动调用的，所以需要在子类中重新定义类的初始化方法。当我们去调用这个类的方法时，

如果在这个类中没有找到，就会判断这个类是否为子类，如果是子类，就会继续到父类中去找。

下面的例子是类的继承，继承上面的Student类
'''
class Graduated(Student):
    
    def __init__(self, name, age, school):     #__是两个下划线！
        self.name = name
        self.age = age
        self.school = school
        
    def dis_school(self):
        print("University of the graduated is:", self.school)
        
graduated = Graduated("wang", 23, "HD")
graduated.dis_school()
graduated.dis_student()
    
```
