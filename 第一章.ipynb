# 第一章        认识神经网络和深度学习
## 神经网络
- 1、神经网络的设计原理是神经元传递信息的方法，神经元由树突、轴突、突触组成，树突接收外界信息，并通过阈值判断是否激活，如果需要激活在通过轴突传递到突触，并传至与它连接的神经元。
- 2、如图1-1为M-P模型，图中的阴影部分为激活函数，w1j、w2j、w3j为权重值。
    $$图1-1 M-P模型$$
    由此可以知道M-P模型的数学表达式如下：
    $${y_j} = f(\sum\limits_{i = 1}^n {{w_{ij}}{x_i} - {\theta _j}} )$$
    
- 3、感知机，它是一种二分类线性模型，其数学表达式为：
    $$f(x) = \operatorname{sign} (wx + b)$$
    其中，参数x为输入向量，w为权重，b为偏置，w.x为点积。
        
## 深度学习
- 图像分类
    - 图像分类的思路是将已经标注好的图片输入到我们搭建好的计算机视觉模型中进行训练，然后使用这个模型对没有标注的图片进行类别预测，只要预测的足够准确，就算解决了图片的分类。
- 图像的目标识别（Object Recognition）和语义分割(Semantic Segmentation)
    - 进行目标识别和语义分割前，必须让搭建的模型知道每个类别的特征。它们有一个很大的不同：目标识别将识别的类别用方框框选出来并打上标签；语义分割则会对识别的类别用同样的像素进行标识并打上标签。目标识别的结果如图2-1（a）所示，语音分割的结果如图2-1（b）所示。
